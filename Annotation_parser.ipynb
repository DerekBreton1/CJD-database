```python
#Import packages
import pandas as pd
```

```python
#Initialize file variable - Path to the VCF file
vcf_file = "normalized_samples_ID.tsv"

#Create pd dataframe from tsv
df = pd.read_csv(vcf_file, delimiter='\t')

#Subset the dataframe for ID and ANN columns
df = df[['ID', 'ANN']]

df
```

```python
df['vid'] = df.reset_index().index + 1
df
```

```python
#Explode IDs down to their own row *Duplicates annotations*
#df = (df.set_index(['ANN'])
#   .apply(lambda x: x.str.split(';').explode())
#   .reset_index())  

df
```

```python
#Explode ANNs down to their own row *Duplicates IDs*
df = (df.set_index(['ID','vid'])
   .apply(lambda x: x.str.split(',').explode())
   .reset_index())  

df
```

```python
#Create temporary datafame to hold annotation data
temp_df = df['ANN'].str.split('|', expand=True)
#num_cols_temp = temp_df.shape[1]  # Number of columns in the temporary DataFrame
temp_df
```

```python
#Add column labels to the dataframe
temp_df.columns = ['ALT', 'Annotation', 'Impact', 'Gene_name', 'Gene_ID', 'Feature_type', 'Feature_ID', 'Biotype', 'Rank/total', 'HGVS.c', 'HGVS.p', 'cDNA_pos', 'AA_pos', 'Distance', 'X', 'Y']

#Add annotation data onto the original dataframe
df = pd.concat([df, temp_df], axis=1)

#Drop the original annotation column
df.drop('ANN', axis=1, inplace=True)

df
```

```python
#Running this downloads the dataframe to a csv
df.to_csv('annotations.csv', index=False)
```
